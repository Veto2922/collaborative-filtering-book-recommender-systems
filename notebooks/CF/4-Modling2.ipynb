# importing


```python
import pandas as pd
import joblib
import numpy as np
import pickle

from scipy.sparse import csr_matrix
# Python v  = 3.7.16
from surprise import SVD, Dataset, Reader, BaselineOnly, AlgoBase , SVDpp
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
user_item_matrix = joblib.load(INPUT)
```

```python
user_item_matrix
```

## Convert NaN values to 0 for Surprise library compatibility


```python
user_item_matrix = user_item_matrix.fillna(0)
```

```python
user_item_matrix
```

## Create a Surprise Dataset


```python
# Create a Surprise Dataset
reader = Reader(rating_scale=(0, 10))  # ratings are on a scale from 0 to 10
data = Dataset.load_from_df(user_item_matrix.stack().reset_index(), reader)
```

## Split the dataset into training and testing sets


```python
trainset, testset = train_test_split(data, test_size=0.1, random_state=42)
```

```python
trainset
```

```python
# testset
```

# 1)Baseline estimate

First, we must make a baseline model (a simple model) to compare it with main model


```python
baseline_model = BaselineOnly()
baseline_model.fit(trainset)
```

## Evaluate the baseline model


```python
predictions = baseline_model.test(testset)
predictions
```

```python
rmse = accuracy.rmse(predictions)
print(f'RMSE: {rmse}')
```

# 2) main model (ALS(FUNK SVD))


```python
# Specify the number of factors for the model (We can make grid search CV to tuning it )
n_factors = 50
model = SVD(n_factors=n_factors, random_state=42 , n_epochs = 10 , biased = False ,  lr_all=0.01)
model.fit(trainset)
```

```python
model.pu.shape  # The user factors
```

```python
model.qi.shape  # The item factors
```

## Evaluate the main model


```python
predictions = model.test(testset)

# Evaluate the model using RMSE (Root Mean Squared Error)
rmse = accuracy.rmse(predictions)
# print(f'RMSE: {rmse}')
```

Well, there is not much difference between the main model and the baseline model , But the real performance shows after deploying and monitoring its performance over time and calculating the hit rate , the coaltitive hit rate, serendipity, and other metrics.
