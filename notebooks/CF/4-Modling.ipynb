```python
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming you have user_item_matrix_scr as your actual user-item matrix and the baseline model is already trained

# Generate predictions using the baseline model
# Example: Assuming your model has a recommend method that generates recommendations for a given user
# You need to iterate over all users and items to generate predictions for all user-item pairs

predicted_ratings = []

for user_id in range(user_item_matrix_scr.shape[0]):
    recommendations = baseline_model.recommend(
        user_id, user_item_matrix_scr[user_id])
    for item_id in range(user_item_matrix_scr.shape[1]):
        # Assuming the recommend method returns a predicted rating for the given user-item pair
        predicted_rating = baseline_model.predict(user_id, item_id)
        predicted_ratings.append(predicted_rating)

# Flatten the actual ratings and predicted ratings for comparison
actual_ratings_flat = user_item_matrix_scr.data
predicted_ratings_flat = np.array(predicted_ratings)

# Compute RMSE
rmse = np.sqrt(mean_squared_error(actual_ratings_flat, predicted_ratings_flat))
print("Root Mean Squared Error (RMSE):", rmse)
```
