# Feature Engineering


```python
CLEAND_BOOKS = r'..\..\data\processed\cleaned_books.csv'
CLEAND_Rating = r'..\..\data\processed\cleaned_Rating.csv'
CLEAND_Users = r'..\..\data\processed\cleaned_users.csv'

FINAL_DF = r'..\..\data\processed\For content base\Final_content_base.csv'
```

# importing


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
books_df = pd.read_csv(CLEAND_BOOKS)

rating_df = pd.read_csv(CLEAND_Rating)

users_df = pd.read_csv(CLEAND_Users)
```

```python
rating_df
```

# First let's get average and number of rating for each book , so we can measure the book popularity in our rank based recommendation system


```python
avg_books_rating = rating_df.groupby(
    'ISBN')['Book-Rating'].mean().reset_index()
```

```python
avg_books_rating
```

```python
avg_books_rating['Rating_count'] = rating_df.groupby(
    'ISBN')['Book-Rating'].count().reset_index()['Book-Rating']
```

```python
avg_books_rating = avg_books_rating.rename(columns={
    'Book-Rating': 'Avg_rating'
})
```

# Now we will merge that with books df


```python
books_df = books_df.merge(avg_books_rating, on='ISBN')
```

```python
books_df.sort_values(by='Rating_count', ascending=False)
```

```python
books_df.columns
```

```python
books_df = books_df[['ISBN', 'Title', 'Author', 'Publication_year',
                     'Publisher', 'Avg_rating', 'Rating_count', 'Book_image']]
```
